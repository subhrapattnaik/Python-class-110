```python
import random
import plotly.express as px
import plotly.figure_factory as ff
import statistics
import csv
import pandas as pd
import plotly.graph_objects as go
```

```python
df=pd.read_csv('data.csv')
data=df["temp"].tolist()
populationMean=statistics.mean(data)
fig=ff.create_distplot([data],["temp"],show_hist=False)
#fig.add_trace(go.Scatter(x=[populationMean,populationMean],y=[0,1],mode="lines",mean="MEAN"))
fig.show()
print("population mean ",populationMean )
print("population standard devision",statistics.stdev(data) )
```

```python
# let's take sample of 100 data points and find  the mean and standard deviation
dataset=[]
for i in range (0,100):
  random_index=random.randint(0,len(data))
  value=data[random_index]
  dataset.append(value)

meanofsample=statistics.mean(dataset)
standardDeviation=statistics.stdev(dataset)

print("mean of sample ", meanofsample )
print("standard Deviation of sample",standardDeviation )
```

```python
# let's calcuate the mean of 100 random data point 1000 times and plot it on the graph
# function to get the mean of given data sample
def random_set_of_mean(counter):
  dataset=[]
  for i in range (0,100):
    random_index=random.randint(0,len(data)-1)
    value=data[random_index]
    dataset.append(value)

  meanofsample=statistics.mean(dataset)
  return meanofsample

def setup():
  mean_list=[]
  for i in range(0,1000):
    set_of_means=random_set_of_mean(100)
    mean_list.append(set_of_means)
  samplingMean=statistics.mean(mean_list)
  sampleingstdev=statistics.stdev(mean_list)
  print("mean of sampling distribution: ",samplingMean)
  print("standard deviation of sampling distribution: ",sampleingstdev)
    
  fig=ff.create_distplot([mean_list],["temp"],show_hist=False)
  #fig.add_trace(go.Scatter(x=[samplingMean,samplingMean],y=[0,1],mode="lines",mean="MEAN"))
  fig.show()  
  
setup()
```
